```python
import requests
def print_links(links):
    for link in links:
        user_input = input("Pressione Enter para ver o próximo link ou digite 'encerrar' para encerrar: ")
        if user_input.lower() == 'encerrar':
            break
        print(link)
url = 'https://api.nasa.gov/mars-photos/api/v1/rovers/opportunity/photos?sol=2010&api_key=CuAHy0N4jOGGfel1p7TAM10KLmRBoBIInehXOn86'
resposta = requests.get(url)
if resposta.status_code == 200:
  fotos = resposta.json()['photos']
  links = [foto['img_src'] for foto in fotos]
  print_links(links)
else:
    print(f"Erro no pedido: {resposta.status_code}")



```
